
This program is for resizing images from initial dataset and save resized images in another folder with folder structure as same as initial dataset.



step 1: connect to google drive for reading images.

```
# connect to google drive
from google.colab import drive
drive.mount('/content/drive', True)
```

step 2: setting path in variables


---


Note: before running this cell ensure that resized_images , resized_images/training and resized_images/test folders created in root of project folder (beside this python file)

```
source_folder = '/content/drive/My Drive/archive/' # end by slash
dest_folder = '/content/drive/My Drive/resized-images/' #end by slash
source_training_folder = source_folder + 'training/' #end by slash
source_test_folder = source_folder + 'test/' #end by slash
dest_training_folder = dest_folder + 'training/' #end by slash
dest_test_folder = dest_folder + 'test/' #end by slash
```

Step 3: resize images of training and test then save them in destination folder.

```
 # resized training images
import os # for reading folders and files address
from PIL import Image # this library for resize and save.
image_size = (32, 32) # size of final image after resize.
```

```
# read folders of training folder
for root, folders, files in os.walk(source_training_folder):
    # loop over each training folder
    for folder in folders:
        # make folder in destination with same name.
        os.mkdir(dest_training_folder+folder)
        # read files of each training folders
        for r2, f2, images in os.walk(source_training_folder+folder):
            # loop over each files in each training folder.
            for image_file_name in images:
                # read each image and resize and save in destination folder.
                img = Image.open(source_training_folder+folder+'/'+image_file_name)
                img = img.resize(image_size)
                img.save(dest_training_folder+folder+'/'+image_file_name) 
```

```
# read folders of test folder
for root, folders, files in os.walk(source_test_folder):
    # loop over each test folder
    for folder in folders:
        print('folder', folder)
        # make folder in destination with same name.
        os.mkdir(dest_test_folder+folder)
        # read files of each test folders
        for r2, f2, images in os.walk(source_test_folder+folder):
            # loop over each files in each test folder.
            for image_file_name in images:
                # read each image and resize and save in destination folder.
                img = Image.open(source_test_folder+folder+'/'+image_file_name)
                img = img.resize(image_size)
                img.save(dest_test_folder+folder+'/'+image_file_name) 
```
